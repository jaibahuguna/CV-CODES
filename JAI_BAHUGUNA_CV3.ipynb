**Task 1:**  Perform Erosion on Fig such that all balls get separated from each other. Optional (you can further apply your connected component analysis algorithm to count total number of balls present in this image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage.io import imread
from skimage.io import imshow
```

```python
img1=cv2.imread('/content/balls.jpg')
print(img1)
```

```python
ret, mask=cv2.threshold(img1, 100, 255, cv2.THRESH_BINARY)
kernal=np.ones((5,5), np.uint8)
# cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
```

```python
dilation=cv2.dilate(mask, kernal, iterations=1)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Dilated Image')
```

```python
erosion=cv2.erode(mask, kernal, iterations=4)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(erosion,cmap='gray')
plt.title('Eroded Image')
```

```python
#Erode-> dilate to get proper shape
# closing=cv2.morphologyEx(erosion, cv2.MORPH_CLOSE, kernal)
dilation2=cv2.dilate(erosion, kernal, iterations=1)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(dilation2, cmap='gray')
plt.title('Dilation of Eroded Image')
```

```python
opening=cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(opening, cmap='gray')
plt.title('Opened Image')
```

```python
closing=cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(closing, cmap='gray')
plt.title('Closed Image')
```

```python
mg=cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(mg, cmap='gray')
plt.title('Gradient Image')
```

```python
th=cv2.morphologyEx(img1, cv2.MORPH_TOPHAT, kernal)
plt.imshow(th, cmap='gray')
```

```python
brain = cv2.imread('/content/brain.png', cv2.IMREAD_GRAYSCALE)

kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(brain, kernel, iterations = 1)
erosion = cv2.erode(dilation, kernel, iterations = 1)
gradient = cv2.morphologyEx(brain, cv2.MORPH_GRADIENT, kernel)
final = cv2.subtract(dilation, erosion)

plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 1)
plt.imshow(brain, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 5, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Dilated Image')

plt.subplot(1, 5, 3)
plt.imshow(erosion, cmap='gray')
plt.title('Eroded Image')

plt.subplot(1, 5, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Gradient Image')

plt.subplot(1, 5, 5)
plt.imshow(final, cmap='gray')
plt.title('Final Image')
```
